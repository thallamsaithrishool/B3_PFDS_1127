**SEABORN**

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset("tips")
sns.boxplot(x="day",y="total_bill",data=tips)
plt.xlabel("Day of the week")
plt.ylabel("Total Bill ($)")
plt.show()
```

```python
student=(["raju","ravi","kiran"])
marks=([23,45,89])
plt.title("Student Marks")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.grid(True)
plt.bar(student, marks)
plt.show()
```

```python
import matplotlib.pyplot as plt
days=["Monday","Tuesday","Wednesday","Thursday","Friday","saturday"]
temp=['15','25','25','30','40','43']
plt.plot(days, temp, marker='o', linestyle='--')
plt.title("Daily Temperature")
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.show()
```

**HEATMAPS**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data=np.random.rand(5,5)
df=pd.DataFrame(data,columns=["A","B","C","D","E"])
corr=df.corr()
sns.heatmap(corr,annot=True,cmap="coolwarm",linewidth=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

**LAB PROGRAM 1**

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('world_population.csv')
d = df[['Country/Territory', '2022 Population']].sort_values(by='2022 Population', ascending=False).head(5)
d.columns = ['c', 'p']

def f(x, pos):
    if x >= 1e9:
        return '{:1.1f}B'.format(x/1e9)
    else:
        return '{:1.0f}M'.format(x/1e6)

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(d['c'], d['p'], color='skyblue')
ax.set_title('Top 5 Countries by 2022 Population')
ax.set_xlabel('Country/Territory')
ax.set_ylabel('2022 Population')
ax.yaxis.set_major_formatter(plt.FuncFormatter(f))
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```

**LAB PROGRAM 2**

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("imdb_top_1000.csv")

genre_counts = df['Genre'].str.split(', ', expand=True).stack().value_counts()

plt.figure(figsize=(12, 6))
plt.bar(genre_counts.index, genre_counts.values, color='#4682B4')

plt.title('Number of Movies Produced by Genre (IMDB Top 1000)', fontsize=16, fontweight='bold')
plt.xlabel('Genre', fontsize=14)
plt.ylabel('Number of Movies', fontsize=14)

plt.xticks(rotation=45, ha='right')

plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.savefig('genre_movie_count_bar_chart_clean.png')
```
