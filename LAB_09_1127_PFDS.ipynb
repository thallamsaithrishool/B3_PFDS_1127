**HEATMAP**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data=np.random.rand(5,5)
df=pd.DataFrame(data,columns=["A","B","C","D","E"])
corr=df.corr()
sns.heatmap(corr,annot=True,cmap="coolwarm",linewidth=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

**PAIRPLOT**

```python
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset("iris")
sns.pairplot(df,hue="species",diag_kind="kde",palette="Set1")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load data
tips=sns.load_dataset("tips")
```

```python
#Barplot
plt.figure(figsize=(8,6))
sns.barplot(x="day",y="total_bill",data=tips,palette="muted")
plt.title("Total Bill by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill")
plt.show()
```

```python
#Boxplot
plt.figure(figsize=(8,6))
sns.boxplot(x="day",y="total_bill",data=tips,palette="Set1")
plt.title("Distribution of Total Bill by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill")
plt.show()
```

```python
#countplot
plt.figure(figsize=(8,6))
sns.countplot(x="day",data=tips,palette="Set2")
plt.title("Number of Appointments by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Number of Appointments")
```

```python
#violin
plt.figure(figsize=(8,6))
sns.violinplot(x="day",y="total_bill",data=tips,palette="Set3")
plt.title("Distribution of Total Bill by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("/content/housing.csv")
data = data.select_dtypes(include=["float64", "int64"])
sns.set(style="whitegrid", palette="coolwarm")
sns.pairplot(data, diag_kind="hist", corner=True)
plt.suptitle("Pair Plot - Relationship Between House Price, Area, and Rooms", y=1.02)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("Titanic-Dataset.csv")
data = data.dropna(subset=["Age", "Fare", "Pclass", "Survived"])
sns.set(style="whitegrid", palette="coolwarm")
sns.pairplot(data, vars=["Age", "Fare", "Pclass"], hue="Survived", diag_kind="kde")
plt.suptitle("Pair Plot - Titanic Passenger Patterns (Age, Fare, Class vs Survival)", y=1.02)
plt.show()
```
